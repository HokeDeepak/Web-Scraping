```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://fastestlaps.com/tracks/adm-miachkovo"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
table=soup.find("table")
```

```python
tbody=table.find("tbody")
```

```python
rows=tbody.find_all("tr")
```

```python
laps=[]
for i in rows:
  col=i.find_all("td")
  laps_info=[c.text for c in col ]
  laps.append(laps_info)
print(laps)
```

```python
Rank=[]
Vehicle=[]
Driver=[]
Time=[]
PSKG=[]
```

```python
for i in laps:
  Rank.append(i[0])
  Vehicle.append(i[1])
  Driver.append(i[2])
  Time.append(i[3])
  PSKG.append(i[4])

print(Rank)
print(Vehicle)
print(Driver)
print(Time)
print(PSKG)
```

```python
data={
    "Rank":Rank,
    "Vehicle":Vehicle,
    "Driver":Driver,
    "Time":Time,
    "PS / KG":PSKG,    
}
```

```python
imp=pd.DataFrame(data)
imp
```

```python
imp.to_csv("ADM Miachkovo lap times.csv",index=False)
```
